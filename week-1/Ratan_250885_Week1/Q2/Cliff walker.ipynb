```python

# 0=up,1=right,2=down,3=left
class Agent():
        def __init__(self,epsilon,decayr,lr,win_reward):
            self.epsilon=epsilon
            self.decayr=decayr
            self.lr=lr
            self.q_table = np.zeros((37, 4))
            self.win_reward=win_reward

        def get_action_train(self,state):
            if random.random() < self.epsilon :
                return random.randint(0, 3)
            else:
                max_val=np.max(self.q_table[state])
                max_indices = np.flatnonzero(self.q_table[state] == max_val) 
                return np.random.choice(max_indices)
            
        def get_action_true(self,state):
            max_val=np.max(self.q_table[state])
            max_indices = np.flatnonzero(self.q_table[state] == max_val) 
            return np.random.choice(max_indices)
        
        def update_q_table (self,state,next_state, action,reward):
            if next_state == 47:
                self.q_table[state,action]=self.lr*(self.win_reward + reward) + (1-self.lr)*self.q_table[state,action]
            else:
                self.q_table[state,action]=self.lr*(self.q_table[next_state,action] + reward) + (1-self.lr)*self.q_table[state,action]
    
        def train_agent(self,episodes):
            env = gym.make("CliffWalking-v1", max_episode_steps=1000)
            observation, info = env.reset()
            
            # Reset the environment to generate the first observation
            
            reward_per_ep=[]
            total_reward_per_ep=0
            while episodes :
                # this is where you would insert your policy
                action = self.get_action_train(observation)
                
                # step (transition) through the environment with the action
                # receiving the next observation, reward and if the episode has terminated or truncated
                new_observation, reward, terminated, truncated, info = env.step(action)
                total_reward_per_ep+=reward
                self.update_q_table(observation,new_observation,action,reward)
                observation=new_observation
                '''
                if reward !=-1 :
                    print (reward,observation)
                '''
                # If the episode has ended then we can reset to start a new episode
                if terminated or truncated:
                    observation, info = env.reset()
                    episodes-=1
                    if self.epsilon*self.decayr>0.01:
                        self.epsilon*=self.decayr
                    if terminated :
                        total_reward_per_ep+=self.win_reward
                    reward_per_ep.append(total_reward_per_ep)
                    total_reward_per_ep=0


            env.close()
            return reward_per_ep
        def test_agent(self,episodes):
            env = gym.make("CliffWalking-v1" ,max_episode_steps=1000)
            observation, info = env.reset()
            # Reset the environment to generate the first observation
            reward_per_ep=[]
            total_reward_per_ep=0
            while episodes :
                # this is where you would insert your policy
                action = self.get_action_true(observation)
                # step (transition) through the environment with the action
                # receiving the next observation, reward and if the episode has terminated or truncated
                observation, reward, terminated, truncated, info = env.step(action)
                total_reward_per_ep+=reward
                '''
                if reward !=-1 :
                    print (reward,observation)
                '''
                # If the episode has ended then we can reset to start a new episode
                if terminated or truncated:
                    observation, info = env.reset()
                    episodes-=1
                    if terminated :
                        total_reward_per_ep+=self.win_reward
                    reward_per_ep.append(total_reward_per_ep)
                    total_reward_per_ep=0

            env.close()
            return reward_per_ep
```

```python
import matplotlib.pyplot as plt
```

```python
import statistics
def draw_graph(count,count1=20,count2=75):
    l=[]
    l1=[]
    l2=[]
    for i in range(count1):
        model1.train_agent(count2)
        result=model1.test_agent(3)
        l.append(statistics.mean(result)/100)
    for i in range(count1):
        model2.train_agent(count2)
        result1=model2.test_agent(3)
        l1.append(statistics.mean(result1)/100)
    for i in range(count1):
        model3.train_agent(count2)
        result2=model3.test_agent(3)
        l2.append(statistics.mean(result2)/100)
    a=np.arange(count1)
    a+=(1+count1*(count-1))
    a*=count2
    plt.plot(a,l,color="red",label="Constant_exp=0.5")
    plt.plot(a,l1,color="blue",label="Constant_exp=0.05")
    plt.plot(a,l2,color="black",label="Exponential_Decay")
    plt.xlabel("Test_Results/100")
    plt.ylabel("Avg_Reward/100")
    plt.title(f"Reward_Graph{count}")
    plt.legend()
    plt.show()
```

```python
model1=Agent(epsilon=0.5,decayr=1,lr=0.1,win_reward=2000)
model2=Agent(epsilon=0.05,decayr=1,lr=0.1,win_reward=2000)
model3=Agent(epsilon=1,decayr=0.95,lr=0.1,win_reward=2000)


draw_graph(1)
draw_graph(2)
draw_graph(3)
draw_graph(4)
```
